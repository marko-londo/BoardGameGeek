```python
import pandas as pd
```

```python
# Read in review data

df = pd.read_csv("../Resources/bgg-19m-reviews.csv", index_col = 0)
df.head()
```

```python
# Drop unnecessary columns

df = df.drop(['comment'], axis=1)
```

```python
df
```

```python
# filter for reviews by users with at least 20 reviews

counts = df['user'].value_counts() 
valids = counts[counts>20].index

df1 = df[df['user'].isin(valids)]

df1.groupby(['user']).count()
```

```python
df1
```

```python
# check for the number of games still reached by our dataset

df1.groupby(['ID']).count()
```

```python
games = df1['ID'].unique()
users = df1['user'].unique()

ind = 1
dic1 = dict()

for user in users:
    dic1[user] = ind
    ind += 1

dic1
```

```python
df1['uID'] = df1['user'].map(dic1)
df1.drop(columns=['user'], inplace=True)
df1
```

```python
# Save a snapshot of the numeric reviews by reviewer

df1.to_pickle('../BoardGameGeek/data_prepped/scores_for_ALS.pk1')
```
