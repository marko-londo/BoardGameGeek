```python
import pandas as pd
```

```python
df = pd.read_pickle('../BoardGameGeek/data_prepped/pivot_by_game.pk1')
df.fillna(0, inplace=True)
sdf = df.astype(pd.SparseDtype(int, fill_value=0))
del df
sdf
```

```python
# Concatenate in one segment at a time in case of memory/address overflow
df2 = pd.read_pickle('../BoardGameGeek/data_prepped/pivot_by_game2.pk1')
df2.fillna(0, inplace=True)
sdf2 = df2.astype(pd.SparseDtype(int, fill_value=0))
del df2
sdf = pd.concat([sdf, sdf2], axis=0)
sdf
```

```python
# Save a snapshot of the pivot by game

sdf.to_pickle('../BoardGameGeek/data_prepped/pivot_by_game_conc.pk1')
```

```python
# Concatenate in one segment at a time in case of memory/address overflow
df2 = pd.read_pickle('../BoardGameGeek/data_prepped/pivot_by_game3.pk1')
df2.fillna(0, inplace=True)
sdf2 = df2.astype(pd.SparseDtype(int, fill_value=0))
del df2
sdf = pd.concat([sdf, sdf2], axis=0)
sdf
```

```python
# Post-process and save a snapshot of the pivot by game
del sdf2
sdf.fillna(0, inplace=True)
sdf = sdf.astype(pd.SparseDtype(int, fill_value=0))
sdf.to_pickle('../BoardGameGeek/data_prepped/pivot_by_game_conc2.pk1')
```

```python
from sklearn.neighbors import NearestNeighbors
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(sdf.values)
distances, indices = knn.kneighbors(sdf.values, n_neighbors=5)
```
